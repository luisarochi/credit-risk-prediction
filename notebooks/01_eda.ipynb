```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuración para visualizaciones
sns.set(style="whitegrid")
plt.style.use('ggplot')
```

```python
# Cargar el dataset
df = pd.read_csv('Credit Risk Benchmark Dataset.csv')
```

```python
# Mostrar primeras filas
df.head()
```

```python
# Tamaño del DataFrame
print("Tamaño:", df.shape)
```

```python
# Información general
df.info()
```

```python
# Estadísticas descriptivas
df.describe()
```

```python
# Verificar valores nulos
df.isnull().sum()
```

```python
# Visualización rápida de valores nulos
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Mapa de calor de valores nulos")
plt.show()
```

```python
sns.countplot(x='dlq_2yrs', data=df)
plt.title('Distribución de Morosidad (dlq_2yrs)')
plt.xlabel('¿Morosidad en 2 años? (0 = No, 1 = Sí)')
plt.ylabel('Cantidad de personas')
plt.show()
```

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.show()
```

```python
# Ingresos vs Morosidad
sns.boxplot(x='dlq_2yrs', y='monthly_inc', data=df)
plt.title('Ingresos mensuales por morosidad')
plt.show()

# Utilización de crédito vs Morosidad
sns.boxplot(x='dlq_2yrs', y='rev_util', data=df)
plt.title('Utilización de crédito por morosidad')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matriz de correlación")
plt.show()
```
